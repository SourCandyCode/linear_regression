```python

import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(20.0,10.0)
```

```python
#read data
data=pd.read_csv('headbrain.csv')
print(data.shape)
data.head()
```

```python
x=data['Head Size(cm^3)'].values
y=data['Brain Weight(grams)'].values
```

```python
#mean dyal x o y
mean_x=np.mean(x)
mean_y=np.mean (y)
#total num dyal values
n=len(x)
#hisab b1 o b2
numer=0

denom=0
for i in range (n):
    numer+=(x[i]-mean_x)*(y[i]-mean_y)
    denom+=(x[i]-mean_x)**2
b1=numer/denom
b=mean_y-(b1*mean_x)
print(b1 ,b)
```

```python
#regression
max_x=np.max(x)+100
min_x=np.min(x)-100
x=np.linspace(min_x, max_x)
y=b+b1* x
plt.plot(x,y,color='#58b970',label='regression line')
plt.scatter(x,y,color='#ef5423',label='scater plot')

plt.xlabel("headsize in cm3")
plt.ylabel("brain weight in grams")

plt.show()
```
